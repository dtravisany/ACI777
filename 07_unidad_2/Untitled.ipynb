```python
import pandas as pd
import folium as maps
```

```python
df=pd.read_json('https://raw.githubusercontent.com/2x3-la/geo-chile/master/chile.json')
```

```python
mapa =  maps.Map(location=[df.lat.mean(),df.lng.mean()],zoom_start=4,control_scale=True)
```

```python
mapa
```

```python
for index, valores in df.iterrows():
    folium.Marker([valores['lat'],valores['lng']],valores['name']).add_to(mapa)
```
