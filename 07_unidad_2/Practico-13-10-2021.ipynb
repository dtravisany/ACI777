#### Vamos a crear un mapa que tiene como centro la ciudad de Santiago de Chile con las coordenadas obtenidas en el [siguiente link](https://es.db-city.com/Chile--Santiago--de-Santiago--Santiago) utilizando folium

### Agregar un marcador para cada comuna  
  
Utilizaremos los datos de latitud y longitud de el siguiente repositorio de (github)[https://github.com/2x3-la/geo-chile]

```python
df= pd.read_json("https://raw.githubusercontent.com/2x3-la/geo-chile/master/chile.json")
```

```python
df
```

```python
df.shape
```

```python
df.describe()
```

```python
df.lat.mean()
```

```python
df.lng.mean()
```

```python
mapa = mapas.Map(location=[df.lat.mean(),df.lng.mean()],zoom_start=4,control_scale=True)
mapa
```

##### Vamos a poner un marcador para cada una de las 345 comunas

```python
for indice, valores in df.iterrows():
    print(valores['name'])
```

```python
df['name']
```

```python
for indice, valores in df.iterrows():
    mapas.Marker([valores['lat'],valores['lng']],valores['name']).add_to(mapa)
```

```python
for indice, valores in df.iterrows():
    mapas.Marker(location=[valores['lat'],valores['lng']],popup=valores['name']).add_to(mapa)
```
