# Data Cleaning

### Missing Data / Datos Faltantes

Este es uno de los problemas más comunes y más difíciles. Generalmente los modelos en los que utilizaremos los datos pueden enfrentar o lidiar con varios tipos de problemas de datos, sin embargo, no pueden resolver el problema de los datos faltantes.

```python
df.describe()
```

Si bien uno puede observar algunas columnas con datos missing, por la fila conteo en `df.describe()` no tenemos una visión global. Por lo que veremos algunas técnicas para visualizar los missing data.


# Técnica 1: El Heatmap

Los _heatmap_ o mapas de calor nos van a mostrar que columnas tienen valores missing. Vamos a seleccionar dos colores, por ejemplo 'azul' y 'rojo', cuando falté un valor la celda (vivienda,feature) tomará el color rojo y cuando esté el valor será azul.





`is null` corresponde a la pregunta ¿está vacío?  

False = no esta vacío (0)
True  = está vacío (1)
La color palette tomara valores cero cuando sea false  y uno cuando  sea true.

```python
df[cols].isnull()
```

```python
cols= df.columns[:30]
colores=['royalblue','darkred']
sns.heatmap(df[cols].isnull(), cmap=sns.color_palette(colores))


```

Podemos ver que existe un patron de datos faltantes en las características.
El eje horizontal de nuestro heatmap muestra el nombre de la función y el eje vertical muestra el número de observación. El rojo representa _missing data_ y el azul representa lo contrario.

Por ejemplo, en la característica `life_sq` presenta datos vacíos en muchas de sus observaciones. Por otro lado `floor` presenta muy pocas celdas vacías. Hay otras características como `timestamp` que siempre tienen datos.

# Técnica 2: Los porcentajes


Si tenemos un conjunto de datos muy grande y la visualización no la podemos realizar, uno puede utilizar la técnica de los porcentajes. Calcularemos el porcentaje de observaciones vacias en cada característica.

```python
for caracteristica in df.columns:
    faltantes = np.mean(df[caracteristica].isnull())
    print('{} -{}%'.format(caracteristica, round(faltantes*100)))
```

Podemos ver la lista que muestra el porcentaje de missing values para cada una de las características. Específicamente, emos que la caracteristica _build_year_ tiene el 45% de sus observaciones vacío, mientras que _floor_ le falta tan solo el 1%.

Esta lista nos permite resumir la información complementando o reemplazando el heatmap.

# Técnica 3: El histograma de los datos perdidos

El _histograma de los datos perdidos_ es una técnica que se utiliza cuando tenemos muchas características. Esto nos permite visualizar patrones y comparar caracteristicas visualmente.


```python
for col in df.columns:
    perdidos = df[col].isnull()
    num_missing = np.sum(perdidos)
    
    if num_missing > 0:
        print('Generando un indicador para: {}'.format(col))
        df['{}_ismissing'.format(col)]= perdidos

        
#Basado en el feature, vamos a graficar el histograma de los datos perdidos

col_perdida= [caracteristica for caracteristica in df.columns if 'ismissing' in caracteristica]
df['num_missing'] = df[col_perdida].sum(axis=1)

df['num_missing'].value_counts().reset_index().sort_values(by='index').plot.bar(x='index', y='num_missing')




```

# ¿Qué hacer?

__No__ existe una solución acordada para enfrentar el problema de los datos vacíos. 

Lo mejor que puede hacer un  _data scientist_ es __estudiar__ la _característica específica_  el _conjunto de datos_ para __decidir__ la mejor manera de manejarlos.

Hay cuatro métodos comunes para manejar los datos faltantes, pero, si la situación es más complicada, debemos ser __creativos__ para utilizar métodos más sofisticados, como el __modelamiento__ de datos faltantes.
