```python
total = "Total"
personas={}
with open('censo.csv.txt') as censo:
    linea = csv.reader(censo, delimiter=',')
    bandera=1
    contador=1
    for columna in linea:
        if bandera==1:
            for elemento in columna:
                print (elemento)
                bandera=0
        else:
            if total in columna[7]:
                print("Eliminando linea de la región "+columna[1]+" porque contiene total")
            else:    
                try:
                    personas[columna[1]]['Hombres']=personas[columna[1]]['Hombres']+int(columna[8])
                    personas[columna[1]]['Mujeres']=personas[columna[1]]['Mujeres']+int(columna[9])
                except KeyError:
                    personas[columna[1]]={'Hombres':int(columna[8]),'Mujeres':int(columna[9])}

                
```
